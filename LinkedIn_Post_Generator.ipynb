```python
from langchain_community.chat_models import ChatOllama
from langchain_core.prompts import PromptTemplate

# Initialize the LLM
llm = ChatOllama(
    model="gemma3:4b",
    temperature=0.7
)

# Prompt template
LINKEDIN_POST_PROMPT = PromptTemplate(
    input_variables=["topic", "language", "tone", "tone_description"],
    template="""
ROLE:
You are a professional LinkedIn content writer.

TASK:
Write a LinkedIn post based on the given topic.

INPUTS:
- Topic: {topic}
- Language: {language}
- Tone: {tone}

TONE DESCRIPTION:
{tone_description}

REQUIREMENTS:
- Platform: LinkedIn
- Length: 2 to 4 short paragraphs
- Structure:
  1. Strong hook
  2. Main insight
  3. Practical reflection or value
  4. Soft conclusion
- Avoid emojis
- Use at most 3 hashtags (optional)

OUTPUT:
Only output the LinkedIn post text.
"""
)
```

```python
linkedin_chain = LINKEDIN_POST_PROMPT | llm

TONE_GUIDELINES = {
    "Professional": "Formal and professional tone, suitable for LinkedIn.",
    "Inspirational": "Motivational, uplifting, encourages action.",
    "Storytelling": "Narrative style, tells a story or journey.",
    "Data-driven": "Includes facts, statistics, or analytical reasoning.",
    "Casual": "Friendly, approachable, slightly informal."
}

def generate_linkedin_post(topic: str, language: str, tone: str) -> str:
    if tone not in TONE_GUIDELINES:
        raise ValueError(f"Invalid tone. Choose from: {list(TONE_GUIDELINES.keys())}")

    result = linkedin_chain.invoke({
        "topic": topic,
        "language": language,
        "tone": tone,
        "tone_description": TONE_GUIDELINES[tone]
    })

    return result.content.strip()
```

```python
topic = "AI in Healthcare"
language = "English"
tone = "Professional"

post = generate_linkedin_post(topic, language, tone)
print(post)
```
